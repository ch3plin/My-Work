```python
import joblib

# Load once when the application starts
model = joblib.load("ministry_model.pkl")
vectorizer = joblib.load("tfidf.pkl")


def predict_ministry(text):
    """
    Predict the ministry responsible for a given subject.
    """

    vector = vectorizer.transform([text])

    prediction = model.predict(vector)

    return prediction[0]
```

```python
if __name__ == "__main__":

    query = "Construction of National Highways"

    result = predict_ministry(query)

    print(result)
```
